## Importing Libraries

```python
import pandas as pd
```

```python
df = pd.read_csv("results_1.csv")
```

## Droping unnecessary Columns

```python
df.head()
```

```python
columns_to_keep = ['categoryName', 'createdOn', 'id', 'name', 'price', 'location.locationDescription', 'location.locationLatitude', 'location.locationLongitude']
```

```python
df = df[columns_to_keep]
df.head()
```

```python
df.to_csv('results_final.csv')
```

## URL Generation from ad name, city, and ad-id

```python
df['clean_name'] = df['name'].str.replace(' ', '-').str.replace(',', '-').str.replace('-+', '-', regex=True)
```

```python
df['final_city'] = df['location.locationDescription'].apply(
    lambda x: x.split()[-1] if x.split()[-1] in ['Kathmandu', 'Lalitpur', 'Bhaktapur'] else 'Nepal'
)
```

```python
df['url'] = 'https://hamrobazaar.com/for-sale-house/' + df['clean_name'] + '-in-' + df['final_city'] + '/' + df['id'].astype(str)
```

```python
df['url'] = df['url'].str.replace('--', '-')
```

## Saving URLS to CSV

```python
df.to_csv('result_with_urls.csv', index=False)
```
